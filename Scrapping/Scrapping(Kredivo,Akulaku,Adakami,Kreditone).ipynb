```python
from google_play_scraper import reviews, Sort
import pandas as pd
```

```python
app_id = 'io.silvrr.installment'  #Akulaku
#app_id = 'com.adakami.dana.kredit.pinjaman'  #adakami
#app_id = 'com.finaccel.android' #Kredivo
#app_id = 'com.id.kredi360' #Kreditone
lang = 'id'
country = 'id'
sort_order = Sort.NEWEST  # Urutkan komentar terbaru
total_comments = 500      # Jumlah komentar yang ingin diambil
batch_size = 100          # Jumlah komentar per batch
```

```python
all_reviews = []
continuation_token = None

while len(all_reviews) < total_comments:
    count = min(batch_size, total_comments - len(all_reviews))

    result, continuation_token = reviews(
        app_id,
        lang=lang,
        country=country,
        sort=sort_order,
        count=count,
        continuation_token=continuation_token
    )

    all_reviews.extend(result)

    if continuation_token is None:
        break  # Tidak ada komentar lagi

print(f"Total komentar yang diambil: {len(all_reviews)}")
```

```python
df = pd.DataFrame(all_reviews)
df = df[['userName', 'score', 'content', 'at']]  # Pilih kolom penting
df.head()
```

```python
df.to_csv('data.csv', index=False, encoding='utf-8-sig')
print("Data berhasil disimpan ke 'data.csv'")
```
